## Part 1: Parsing and Dividing Data

```python
# Read all data
allData = list()
with open("iris.txt", "r") as filestream:
  for line in filestream:
    currentline = line.strip().split(",")
    allData.append([float(currentline[0]), float(currentline[1]), float(currentline[2]), float(currentline[3]), currentline[4]])
print(allData)
```

```python
# test data structure
def test_dataset(data):
    if len(data) != 150:
        return False
    
    for row in data:
        if len(row) != 5:
            return False
        
        for column in row[:-1]:
            if type(column) != float:
                return False
            
        if type(row[-1]) != str:
            return False
    
    return True
print(test_dataset(allData))
```

```python
# Split data (80% train so first 40 of each category, 20% test so last 10 of each category)
trainData = allData[0:40] + allData[50:90] + allData[100:140]
print("trainData count:", len(trainData))
testData = allData[40:50] + allData[90:100] + allData[140:150]
print("testData count:", len(testData))
```

## Part 2: Build LDA Classifier

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# separate input and output
train_x = [sublist[0:4] for sublist in trainData]
print(train_x)
train_y= [sublist[4] for sublist in trainData]
print(train_y)
test_x = [sublist[0:4] for sublist in testData]
test_y= [sublist[4] for sublist in testData]
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(train_x, train_y)
```

```python
# Prediction for Train Set
train_pred = lda.predict(train_x)

# Get correct vs incorrect count
train_correct = sum(train_pred[i] == train_y[i] for i in range(len(train_pred)))
train_incorrect = sum(train_pred[i] != train_y[i] for i in range(len(train_pred)))
print(train_correct, train_incorrect)
```

```python
# Prediction for Test Set
test_pred = lda.predict(test_x)
print("Test prediction: \n", test_pred)
test_correct = sum(test_pred[i] == test_y[i] for i in range(len(test_pred)))
test_incorrect = sum(test_pred[i] != test_y[i] for i in range(len(test_pred)))
print(test_correct, test_incorrect)
```

```python
# Report accuracy
print("Training Accuracy:", train_correct/(train_correct+train_incorrect))
print("Testing Accuracy:", test_correct/(test_correct+test_incorrect))
```

## Part 3:  QDA Classifier

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
qda = QuadraticDiscriminantAnalysis()
qda.fit(train_x, train_y)
```

```python
train_pred = qda.predict(train_x)
train_correct = sum(train_pred[i] == train_y[i] for i in range(len(train_pred)))
train_incorrect = sum(train_pred[i] != train_y[i] for i in range(len(train_pred)))

test_pred = qda.predict(test_x)
test_correct = sum(test_pred[i] == test_y[i] for i in range(len(test_pred)))
test_incorrect = sum(test_pred[i] != test_y[i] for i in range(len(test_pred)))

# Report accuracy
print("Training Accuracy:", round(train_correct/(train_correct+train_incorrect), 4))
print("Testing Accuracy:", test_correct/(test_correct+test_incorrect))
```
