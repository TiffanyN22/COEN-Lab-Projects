## Part 1: Adding 2 100x100 Arrays

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
data1 = np.random.randint(1, 11, (100, 100))
data2 = np.random.randint(1, 11, (100, 100))
print(data1)
print(data2)
```

```python
# test adding with + operator
print(data1 + data2)
```

```python
# time addition with + operator
allTimes = np.empty(1000)
dataSum = np.zeros((100, 100))
for i in range(1000):
  startTime = time.time()
  dataSum = data1 + data2
  allTimes[i] = time.time() - startTime

print("Average:", np.average(allTimes))
print("Standard Deviation:", np.std(allTimes))

plt.hist(allTimes, bins=5, range=(np.min(allTimes), np.max(allTimes)))
plt.title("Time for Addition w/ + Operator")
plt.xlabel("Time")
plt.ylabel("Occurances")
plt.show()
```

```python
# test adding with loop
dataSum2 = np.zeros((100, 100))
for i in range(len(data1)):
    for j in range(len(data1[i])):
        dataSum2[i][j] = data1[i][j] + data2[i][j]
print(dataSum2)
```

```python
# time addition with loop
allTimes = np.empty(1000)
dataSum2 = np.zeros((100, 100))
for count in range(1000):
  startTime = time.time()
  for i in range(len(data1)):
    for j in range(len(data1[i])):
        dataSum2[i][j] = data1[i][j] + data2[i][j]
  allTimes[count] = time.time() - startTime

print("Average:", np.average(allTimes))
print("Standard Deviation:", np.std(allTimes))

plt.hist(allTimes)
plt.title("Time for Addition w/ loops")
plt.xlabel("Time")
plt.ylabel("Occurances")
plt.show()
```

## Part 2: Numpy and Matplotlib Program
The purpose of this program is to generate 500 points of random data that is normally distributed and plot a histogram of that data

```python
mean = 5
std = 2
normalDist = np.random.normal(loc=mean, scale=std, size=500) # generate normally distributed data

plt.hist(normalDist, bins=15) # plot histogram

# plot lines for mean and standard deviation range
plt.axvline(x = mean, color = 'r', label = 'mean')
plt.axvline(x = mean-std, color = 'g', label = 'mean-std')
plt.axvline(x = mean+std, color = 'b', label = 'mean+std')

#label plot
plt.title("Normal Distribution Example")
plt.xlabel("Value")
plt.ylabel("Occurances")
plt.legend()
plt.show()
```
